```python
from statsmodels.tsa.arima.model import ARIMA

fig, ax = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(temp, ax=ax[0], lags=20)
plot_pacf(temp, ax=ax[1], lags=20)
plt.show()

# Grid search with information criteria
p_values = range(0, 3)  # Adjust based on ACF plot
d_values = range(0, 2)  # Typically 0 or 1 for differencing
q_values = range(0, 3)  # Adjust based on PACF plot

best_aic = float("inf")
best_params = None

for p, d, q in product(p_values, d_values, q_values):
    try:
        model = ARIMA(temp, order=(p, d, q), enforce_stationarity=False, enforce_invertibility=False)
        results = model.fit()
        aic = results.aic
        if aic < best_aic:
            best_aic = aic
            best_params = (p, d, q)
    except:
        continue

print("Best AIC:", best_aic)
print("Best (p, d, q):", best_params)
```
